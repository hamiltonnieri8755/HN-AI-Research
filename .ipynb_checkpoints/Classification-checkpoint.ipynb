```python
# import libraries
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the data file
df = pd.read_csv('content/diabetes.csv')
df.head()
```

```python
df.describe().T
```

```python
df['Pregnancies'].value_counts()
```

```python
for column in df.columns:
    plt.figure(figsize=(10, 5))
    df[column].hist(bins=30)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Freq')
    plt.show()
```

```python
df_processed = df.copy()
for column in df_processed.columns:
    if column in ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']:
        mean_value = df_processed[column][df_processed[column] != 0].mean()
        org_col_dtype = df_processed[column].dtype
        print(f'{column}-{df_processed[column].dtype}-{mean_value}')
        df_processed[column] = df_processed[column].replace(0, mean_value)
        df_processed[column] = df_processed[column].astype(org_col_dtype)
        print(df_processed[column].dtype)
```

```python
array_org = df.values
X_org = array_org[:,0:8]
Y_org = array_org[:,8]
kfold_org = KFold(n_splits=10, random_state=7, shuffle=True)
model_org = LogisticRegression(solver='liblinear')
results_org = cross_val_score(model_org, X_org, Y_org, cv=kfold_org)
print("Accuracy: %.3f%% (%.3f%%)" % (results_org.mean()*100.0, results_org.std()*100.0))
```

```python
array = df_processed.values
X = array[:,0:8]
Y = array[:,8]
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
model = LogisticRegression(solver='liblinear')
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```
