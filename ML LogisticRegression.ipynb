```python
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
X = np.array([[3.1, 7.2], [4, 6.7], [2.9, 8], [5.1, 4.5], [6, 5],
[5.6, 5], [3.3, 0.4], [3.9, 0.9], [2.8, 1], [0.5, 3.4], [1, 4], [0.6,
4.9]])
print(X)
y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2, 3, 3, 3])
```

```python
# create the logistics regression classifier
classifier = linear_model.LogisticRegression(solver='liblinear', C=1)
classifier.fit(X,y)
```

```python
output = classifier.predict([[4, 6.7]])
print(output)
```

```python
print(X)
```

```python
# min x axis value
min_x, max_x = X[:,0].min() - 2, X[:, 0].max() + 2
print(max_x)
```

```python
# min y axis value
min_y, max_y = X[:,1].min() - 2, X[:, 1].max() + 2
```

```python
# define the mesh grid step size
mesh_step = 0.01

# define the mesh grid of x and y values
x_vals, y_vals = np.meshgrid(np.arange(min_x, max_x, mesh_step), np.arange(min_y, max_y, mesh_step))
```

```python
np.c_[x_vals.ravel(), y_vals.ravel()]
```

```python
mesh_grid_output = classifier.predict(np.c_[x_vals.ravel(), y_vals.ravel()])
print(mesh_grid_output)
print(mesh_grid_output.reshape(x_vals.shape))
print(x_vals)
```

```python
plt.figure()
plt.pcolormesh(x_vals, y_vals, mesh_grid_output.reshape(x_vals.shape), cmap = plt.cm.gray)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hamilton TEST')
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```
